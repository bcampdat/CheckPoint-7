## ***4. Qu es un condicional?***



    Los condicionales nos permiten tomar decisiones, crear ramificaciones en la lgica del software que estamos construyendo: bajo que condiciones debo ejecutar las instrucciones? qu hago si no se cumple mis requisitios?

[**if / else** ](https://www.youtube.com/shorts/Zw4PTaGUUXo?feature=share)

`````python
let edad = 20;

if (edad >= 18) {
    console.log("Eres mayor de edad");
} else {
    console.log("Eres menor de edad");
}
`````

* ***Tipos de condicionales***

|     Smbolo   |             Explicacin         |                Ejemplo           |
|:-------------:|:-------------------------------:|:--------------------------------:|
|       >       |       es mayor que              |         if (1 > 0)               |
|       <       |       es menor que              |         if (1 < 0)               |
|     &&        |           y                     |         if (1 > 0 && 67 > 0)     |
|     ||        |               o                 |         if (1 > 10 || 67 > 0)    |
|     ==        |       es igual en valor         |         if ("3" == 3)            |   
|     ===	    |      es igual en valor y tipo   |         if ("3" === "3")         |
|      !        |          no                     |         if (!(1 > 0))            |
|      !=       |         no es igual             |         if ("Doctor" != "Who")   |
|    !==	    |   no es igual en valor o tipo   |         if ("Doctor" !== "Who")  |
|    >=         |       es mayor o igual que      |         if (10 >= 10)            |
|    <=         |       es menor o igual que      |         if (10 <= 20)            |
|    true       |           Verdad	              |         if (true)                |
|    false      |           Falso                 |         if (false)               |


[**Switch**](https://www.youtube.com/shorts/aF0B2QV1qwA?feature=share)

Se comporta como un condicin cuyo valor es igualado en todos los casos.

````
switch (variable) {
    case 0:
        ...
        break;
    case 1:
        ...
        break;
    case 2:
        ...
        break;
    default:
        ...
        break;
}
````


Veamos un ejemplo.	

`````python
const edad = 65;

switch (edad) {
    case 0:
        console.log("Recin nacido");
        break;
    case 18:
        console.log("ya eres mayor de edad");
        break;
    case 65:
        console.log("Recin jubilado");
        break;
    default:
        console.log("An estas vivo?");
        break;
}

// Recin jubilado
`````
