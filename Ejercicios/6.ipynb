## ***6. Cul es la diferencia entre una declaracin de funcin y una expresin de funcin?***



Javascript tiene dos tipos de funciones, las annimas y las que no son annimas. De ah que se clasifican como function declaration y function expression.

**function declaration**

Se llama as a las funciones creadas usando el keyword function seguido del nombre.

````
function message(){
	// some code
}
````
**function expression**

Se le llama as a las funciones annimas que son almacendas en una variable

````
const message = function(){
	// some code
}
````
**La diferencia?**

La diferencia est en cmo usas las funciones.

Cuando se declara una funcin usando function declaracion el hoisting se encarga de hacerlo disponible a lo largo de todo el contexto de ejecucin. Por eso podemos ejecutar la funciona antes de declararla.

Aprende ms sobre el [Hoisting: ](https://youtu.be/W5MVAWlwwjU)


````
message() // imprime "hola"
function message(){
	console.log("hola")
}
````

Mientras que si hacemos lo mismo usando la forma function expression nos genera un error por el asunto del hoisting
````
message() // ReferenceError: message is not defined
const message = function(){
	console.log("hola")
}
````
Y para evitar el error primero hay que crear la funcin y luego usarla
````
const message = function(){
	console.log("hola")
}
message() // hola
````     

[FUNCIONES / EXPRESIONES ](https://youtu.be/MtT6xcgaMwI)

                                                            EJERCICIOS

1. Observa la diferencia y crea una funcionEjemplo y una espresion de expresionFuncion que impriman "Hola, soy una funcin", "Hola, soy una expresin de funcin" respectivamente

2. Declara una funcion Saludar, 
que toma un nombre como argumento y devuelve un saludo personalizado: 

3. Declara una  expresion de la funcion Saludo, 
que toma un nombre como argumento y devuelve un saludo personalizado: 

4. Aplica el hoisting. Define una funcin llamada respuesta que imprima en la consola el mensaje "Aprendiendo JavaScript".
Llama a la funcin respuesta para que se ejecute y muestre el mensaje en la consola.
Para facilitar la comprension usa respuesta 1, respuesta 2 y respuesta 3.

 [Soluciones :](ej6.js)
